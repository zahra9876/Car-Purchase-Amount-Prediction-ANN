# STEP[0]: LIBRARIES IMPORT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# STEP[1]: IMPOET DATASET

```python
car_df= pd.read_csv('Car_Purchasing_Data.csv', encoding= 'ISO-8859-1')
```

```python
car_df
```

# STEP[2]: VISUALIZE DATASET

```python
sns.pairplot(car_df)
```

```python
X = car_df.drop(['Customer Name','Customer e-mail','Country','Net Worth'], axis=1)
```

```python
X
```

```python
Y = car_df['Car Purchase Amount']
```

```python
Y
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled.shape
```

```python
scaler.data_max_
```

```python
scaler.data_min_
```

```python
Y = Y.values.reshape(-1,1)
```

```python
Y_scaled = scaler.fit_transform(Y)
```

```python
Y_scaled
```

# STEP[4]: TRAINING THE MODEL


```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled,Y_scaled, test_size = 0.25 )
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(40, input_dim = 5, activation = 'relu'))
model.add(Dense(40, activation = 'relu'))
model.add(Dense(1, activation = 'linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer= 'adam' , loss = 'mean_squared_error')
```

 === Summary of model.fit() parameters ===
 X_train, Y_train â†’ training data (inputs and targets)
 
 epochs           â†’ number of times the model sees the entire dataset

 batch_size       â†’ number of samples processed before updating model weights
 
 verbose          â†’ level of training output shown:
                   /  0 = silent
                   /  1 = progress bar (default)
                   /  2 = one line per epoch
                     
 validation_split â†’ fraction of training data used for validation (e.g. 0.2 = 20%)
 
 epochs_hist      â†’ stores the training history (loss, val_loss, accuracy, etc.)

 You can access results like:
 
 epochs_hist.history['loss']       â†’ training loss per epoch
 
epochs_hist.history['val_loss']   â†’ validation loss per epoch

epochs_hist.history['accuracy']   â†’ training accuracy (for classification)

epochs_hist.history['val_accuracy'] â†’ validation accuracy

```python
epochs_hist = model.fit(X_train, Y_train , epochs = 100, batch_size = 25 , verbose = 1, validation_split = 0.2)
```

```python
# === Training results summary ===
# Model: same architecture, different training parameters
# (Final training loss shown for each run)

# epochs = 100, batch_size = 75 â†’ final loss = 1.9031e-05
# epochs = 100, batch_size = 25 â†’ final loss = 4.0217e-06  less time than epochs = 500
# epochs = 20,  batch_size = 25 â†’ final loss = 4.7729e-04
# epochs = 500, batch_size = 25 â†’ final loss = 1.3318e-06

# Observation:
# - Smaller batch_size (25) generally produced lower loss (better fit).
# - Increasing epochs from 20 â†’ 100 significantly reduced loss.
```

# STEP[5]: EVALUATING THE MODEL

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss Progress During Training')
plt.ylabel('Training and Validation Loss')
plt.xlabel('Epoch number')
plt.legend(['Training Loss','Validation Loss'])
```
